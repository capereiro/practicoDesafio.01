```python
def tuples2lists(df):
    for i in range(len(df.index)):
        l = []
        for j in range(len(df.columns)):
            if df.iloc[i,j] != None:
                if type(df.iloc[i,j]) == str:
                    l.append(df.iloc[i,j])
                else:
                    for h in df.iloc[i,j]:
                        if h != "":
                            l.append(h)
        if len(l) == 0:
            df.iloc[i,0] = None
        elif len(l) == 1:
            df.iloc[i,0] = l[0]
        else:
            df.iloc[i,0] = l
    df = df[[0]]
    return df
```
